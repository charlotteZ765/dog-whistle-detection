```python
df_detection = pd.read_csv("df_detection_full.csv")
detection_result_few=pd.read_csv("detection_fewshot_v2.csv")
detection_result=pd.read_csv("dog_whistle_combined v2.csv")
```

```python
detection_result=pd.read_csv("dog_whistle_combined v2.csv")
```

```python
detection_result_few=pd.read_csv("detection_fewshot_v2.csv")
```

```python
detection_result_few
```

```python
columns_to_keep = ['example', 'dog_whistle_present', 'same_meaning', 'cosine_similarity']
detection_result_few = detection_result_few[columns_to_keep]
```

```python
detection_result_few = detection_result_few.rename(
    columns={col: f'llama_3.3_fewshot_{col}' for col in detection_result_few.columns if col != 'example'}
)
```

```python
detection_result_few
```

```python
df_detection_result = pd.merge(df_detection, detection_result_few, on='example', how='inner')
df_detection_result
```

```python
detection_result
```

```python
columns_to_keep = ['original_example', 'dog_whistle_present','same_meaning', 'cosine_similarity']
detection_result = detection_result[columns_to_keep]
```

```python
detection_result = detection_result.rename(
    columns={col: f'llama_3.3_zeroshot_{col}' for col in detection_result.columns if col != 'original_example'}
)
```

```python
detection_result.rename(columns={'original_example': 'example'}, inplace=True)
```

```python
detection_result
```

```python

df_detection_result
```

```python
df_detection_result_joined = pd.merge(df_detection_result, detection_result, on='example', how='inner')
```

```python
df_detection_result_joined
```

```python
df_detection_result_joined['llama_3.3_zeroshot_dog_whistle_present'] = df_detection_result_joined['llama_3.3_zeroshot_dog_whistle_present'].map({
    True: 'coded',
    False: 'noncoded'
})
```

```python
df_detection_result_joined.rename(columns={'example': 'sentence'}, inplace=True)
```

```python
df_detection_result_joined
```

```python
df_detection_result_joined['llama_3.3_fewshot_dog_whistle_present'] = df_detection_result_joined['llama_3.3_fewshot_dog_whistle_present'].replace('non-coded', 'noncoded')

df_detection_result_joined['llama_3.3_zeroshot_dog_whistle_present'] = df_detection_result_joined['llama_3.3_zeroshot_dog_whistle_present'].replace('non-coded', 'noncoded')
df_detection_result_joined['label'] = df_detection_result_joined['label'].replace('non-coded', 'noncoded')
```

```python
df_detection_result_joined = df_detection_result_joined.drop(columns=['idx', 'dog_whistle', 'dog_whistle_root', 'ingroup', 'definition'])
```

```python
df_detection_result_joined.to_csv("llama_3.3_detection_result.csv", index=False)
```
